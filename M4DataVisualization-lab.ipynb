# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
%matplotlib inline

QUERY = """
SELECT * 
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

df['ConvertedComp'].plot(kind='hist')
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Number of Responders')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
df['Age'].plot(kind='box')
plt.title('Box Plot of Age')
plt.ylabel('Values')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.scatter(df['Age'], df['WorkWeekHrs'])
plt.title('Scatter Plot of Age and Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age'], alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""
df_db = pd.read_sql_query(QUERY, conn)
df_db = df_db['DatabaseDesireNextYear'].value_counts().head(5)
labels = df_db.index
sizes = df_db.values
plt.pie(sizes, labels=labels, autopct='%1.2f%%')
plt.title('Top 5 Databases to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df_stacked = pd.read_sql_query(QUERY, conn)
median_weekhrs = df_stacked['WorkWeekHrs'].median()
median_coderev = df_stacked['CodeRevHrs'].median()

df_median = pd.DataFrame({'Median WorkWeekHrs': [median_weekhrs], 'Median CodeRevHrs': [median_coderev]})
df_median=df_stacked.groupby('Age').median()
df_median.plot(kind='bar', stacked=True)
plt.xticks(rotation=0, ha='center')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.legend(bbox_to_anchor=(1.33, 0.5))
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df_line = pd.read_sql_query(QUERY, conn)
median_comp = df_line['ConvertedComp'].median()

df_median_comp = pd.DataFrame({'Median ConvertedComp': [median_comp]})
df_median_comp =df_line.groupby('Age').median()
df_median_comp.plot(kind='line')
plt.xticks(rotation=0, ha='center')
plt.title('Line Chart of Median ConvertedComp for Age Group 45 to 60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.legend(loc='upper left')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""
df_mb = pd.read_sql_query(QUERY, conn)
df_mb=df_mb['MainBranch'].value_counts()
df_mb.plot(kind='barh')
plt.title('Horizontal Bar Chart of Main Branch')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()
```
